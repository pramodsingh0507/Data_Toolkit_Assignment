```python
# Difference between shallow copy and deep copy.
# Shallow Copy in NumPy

# Create an original array
original_array = np.array([1, 2, 3, 4, 5])

# Create a shallow copy using slicing
shallow_copied_array = original_array[:]

# Modify the shallow copied array
shallow_copied_array[0] = 99

print("Original array:", original_array)
print("Shallow copied array:", shallow_copied_array)
```

```python
# Deep Copy in NumPy
# Create an original array
original_array = np.array([1, 2, 3, 4, 5])

# Create a deep copy using the copy() method
deep_copied_array = original_array.copy()

# Modify the deep copied array
deep_copied_array[0] = 99

print("Original array:", original_array)
print("Deep copied array:", deep_copied_array)
```

```python
# Ques-6) Create a 3D NumPy array of shape (3,3,3) containing random integers between (1-10). Perform the following operations:
# a) Find the indices of the maximum values along each depth level (third axis).
# b) Perform element wise multiplication between both arrays


```

```python
# Ans
import numpy as np

# Step 1: Create a 3D array with random integers between 1 and 10
array = np.random.randint(1, 11, size=(3, 3, 3))

print("Original Array:")
print(array)

# Step 2: Find the indices of the maximum values along each depth level (third axis)
# `np.argmax` returns the indices of the maximum values along the specified axis
# To get indices for each depth level (third axis), use axis=2
max_indices = np.argmax(array, axis=2)

print("\nIndices of Maximum Values Along Each Depth Level (third axis):")
print(max_indices)

# Step 3: Perform element-wise multiplication between the original array and the indices array
# The indices array has shape (3, 3) and needs to be expanded to match the shape of the original array (3, 3, 3)
# Use broadcasting to perform element-wise multiplication
# We need to create a new array from max_indices to match the dimensions for multiplication
indices_expanded = max_indices[..., np.newaxis]

# Perform element-wise multiplication
result = array * indices_expanded

print("\nResult of Element-wise Multiplication:")
print(result)
```

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a time-series dataset
# Generate a date range
date_range = pd.date_range(start='2024-01-01', periods=100, freq='D')

# Generate random temperature and humidity data
np.random.seed(0)  # For reproducibility
temperature = np.random.uniform(low=20, high=30, size=len(date_range))  # Temperature between 20°C and 30°C
humidity = np.random.uniform(low=30, high=80, size=len(date_range))  # Humidity between 30% and 80%

# Create the DataFrame
df = pd.DataFrame({
    'Date': date_range,
    'Temperature': temperature,
    'Humidity': humidity
})

# Plotting
fig, ax1 = plt.subplots()

# Plot Temperature on the left y-axis
ax1.plot(df['Date'], df['Temperature'], 'r-', label='Temperature')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color='r')
ax1.tick_params(axis='y', labelcolor='r')

# Create a second y-axis for Humidity
ax2 = ax1.twinx()
ax2.plot(df['Date'], df['Humidity'], 'b-', label='Humidity')
ax2.set_ylabel('Humidity (%)', color='b')
ax2.tick_params(axis='y', labelcolor='b')

# Set the title of the plot
plt.title('Temperature and Humidity Over Time')

# Display the plot
plt.show()
```

```python
# Ques 17 Create a Seaborn scatter plot of two random arrays, colour points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Generate two random arrays
np.random.seed(0)  # For reproducibility
x = np.random.uniform(-10, 10, size=100)
y = np.random.uniform(-10, 10, size=100)

# 2. Determine the quadrant of each point
def determine_quadrant(x, y):
    if x >= 0 and y >= 0:
        return 'Quadrant I'
    elif x < 0 and y >= 0:
        return 'Quadrant II'
    elif x < 0 and y < 0:
        return 'Quadrant III'
    else:
        return 'Quadrant IV'

# Create a DataFrame
df = pd.DataFrame({
    'x': x,
    'y': y
})

# Apply function to determine quadrants
df['Quadrant'] = df.apply(lambda row: determine_quadrant(row['x'], row['y']), axis=1)

# 3. Create the scatter plot using Seaborn
plt.figure(figsize=(10, 6))
scatter_plot = sns.scatterplot(data=df, x='x', y='y', hue='Quadrant', palette='Set1', style='Quadrant')

# 4. Label the axes and set the title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')

# 5. Display the legend and plot
plt.legend(title='Quadrant')
plt.show()
```

```python
import plotly.graph_objects as go
import numpy as np

# 1. Generate random dataset
np.random.seed(0)  # For reproducibility
x = np.linspace(0, 10, 100)  # 100 points from 0 to 10
y = np.random.randn(100)  # Random values from a normal distribution

# 2. Create the line plot
fig = go.Figure()

# Add a line trace
fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='Random Data'))

# 3. Label the axes and set the title
fig.update_layout(
    title='Simple Line Plot',
    xaxis_title='X-axis',
    yaxis_title='Y-axis'
)

# 4. Show the plot
fig.show()
```

```python
# Ques 21 Using Plotly, create an interactive pie chart of randomly generated data, add labels and percentages,
# And set the title as 'Interactive Pie Chart'.

import plotly.graph_objects as go
import numpy as np

# 1. Generate random data
np.random.seed(0)  # For reproducibility
labels = [f'Category {i}' for i in range(5)]
values = np.random.randint(10, 100, size=len(labels))

# 2. Create the pie chart
fig = go.Figure(data=[go.Pie(labels=labels, values=values, textinfo='label+percent', hole=0.3)])  # hole=0.3 for a donut chart, set to 0 for a classic pie chart

# 3. Set the title
fig.update_layout(title='Interactive Pie Chart')

# 4. Show the plot
fig.show()
```
